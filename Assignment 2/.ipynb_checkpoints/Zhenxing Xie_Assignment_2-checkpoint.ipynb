**[20 points] Classification I: Creating a classification algorithm**.

**(a)** Build a working version of a binary kNN classifier using the skeleton code below.

**(b)** Load the datasets to be evaluated here. Each includes training features ($\mathbf{X}$), and test features ($\mathbf{y}$) for both a low dimensional ($p = 2$ features/predictors) and a high dimensional ($p = 100$ features/predictors). For each of these datasets there are $n=100$ observations of each. They can be found in the `data` subfolder in the `assignments` folder on github. Calculate the time it takes to make the predictions in each case and the overall accuracy of each set of test data predictions.

**(d)** Compare your implementation's accuracy and computation time to the scikit learn [KNeighborsClassifier](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) class. How do the results and speed compare?

**(e)** Some supervised learning algorithms are more computationally intensive during training than testing. What are the drawbacks of the prediction process being slow?

**ANSWER**:

```python
# (a) Write your own kNN classifier


class Knn:
# k-Nearest Neighbor class object for classification training and testing
    def __init__(self):        
        self.X_train = None
        self.y_train = None  
 
    def fit(self, x, y):
        # Save the training data to properties of this class
        self.X_train = x
        self.y_train = y
        return self
    
    
    
    def predict(self, x, k):
        
        # Importing the method called "Counter" for the predict function
        
        from collections import Counter
        
        # Variable to store the estimated class label for 
    
        y_hat = []        

        # Looping the observations in the test dataset
        
        for i in x:

            # Calculating the distance from each vector in x to the training data

            distances = [np.sqrt(np.sum((x_train - i)**2)) for x_train in self.X_train] 
            
            # Sorting the distances and returning indexes

            nearest = np.argsort(distances) 
            
            # Returning the categories for the nearest K distances

            topK_y = [self.y_train[j] for j in nearest[:k]] 

            # Counting the numbers for the category labels
            
            votes = Counter(topK_y)
            
            # Picking the category label which appears for the most time for a observation in x
            
            y_hat.append(votes.most_common(1)[0][0]) 

        # Return the estimated targets
        
        return np.array(y_hat)
        
# Metric of overall classification accuracy
#  (a more general function, sklearn.metrics.accuracy_score, is also available)
                
def accuracy(y,y_hat):
    nvalues = len(y)
    accuracy = sum(y == y_hat) / nvalues
    return accuracy
```

```python
# (b) 

# Loading the low-dimensional dataset

x_train_low = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_X_train_low.csv", 
                         delimiter=',')
y_train_low = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_y_train_low.csv", 
                         delimiter=',')
x_test_low = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_X_test_low.csv", 
                        delimiter=',')
y_test_low = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_y_test_low.csv", 
                        delimiter=',')

# Loading the high-dimensional dataset

x_train_high = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_X_train_high.csv", 
                          delimiter=',')
y_train_high = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_y_train_high.csv", 
                          delimiter=',')
x_test_high = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_X_test_high.csv", 
                         delimiter=',')
y_test_high = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_y_test_high.csv", 
                         delimiter=',')
```

```python
# (c) Evaluating the performance of your kNN classifier on a low- and a high-dimensional dataset, 
# and time used and the predictions of each.

import time
k = 5

# Training the classifier on the low-dimensional dataset

start = time.time()
low = Knn()
low.fit(x_train_low, y_train_low)

# Evaluating performance on test set

low_y = low.predict(x_test_low, k)
end = time.time()

print("LOW dimension precision: ", accuracy(y_test_low, low_y))
print("time consumed: %.4f s" % (end - start))

print("-"*80)
# Training the classifier on the high-dimensional dataset

start = time.time()
high = Knn()
high.fit(x_train_high, y_train_high)

# Evaluating performance on test set

high_y = high.predict(x_test_high, k)
end = time.time()
print("HIGH dimension precision: ", accuracy(y_test_high, high_y))
print("time consumed: %.4f s" % (end - start))
```

```python
# (d) Comparing the previous implementation's accuracy and computation time to the scikit learn 
#     KNeighborsClassifier class

from sklearn.neighbors import KNeighborsClassifier

# Training the classifier by using the scikit learn KNeighborsClassifier on the low-dimensional dataset

start = time.time()
low_sl = KNeighborsClassifier(n_neighbors = k)
low_sl.fit(x_train_low, y_train_low)

# Evaluating performance on test set

low_sl_pred = low_sl.predict(x_test_low)
end = time.time()
print("LOW dimension precision: ", accuracy(y_test_low, low_sl_pred))
print("time consumed: %.4f" % (end - start))

print('-'*80)

# Training the classifier by using the scikit learn KNeighborsClassifier on the high-dimensional dataset

start = time.time()
high_sl = KNeighborsClassifier(n_neighbors = k)
high_sl.fit(x_train_high, y_train_high)

# Evaluating performance on test set

high_sl_pred = high_sl.predict(x_test_high)
end = time.time()
print("HIGH dimension precision: ", accuracy(y_test_high, high_sl_pred))
print("time consumed: %.4f" % (end - start))
```

> Comparing the implementation's accuracy and computation time by using the above KNN classifier to its peer by utilizing the scikit learn KNeighborsClassifier class, we would conclude that the scikit learn KNeighborsClassifier class perform better in computational speed and have the same accuracy with the above KNN classifier.

```python
# Creating classification errors for both training set and test set

train_class_err = []
test_class_err = []

# Looping k from number 1 to 500

for k in (range(1, 501)):
    
    # Training models for k = 1,2,...500
    
    model_k = KNeighborsClassifier(n_neighbors = k)
    
    training_fit = model_k.fit(training_set, training_label)
    
    training_pred_label = model_k.predict(training_set)
    
    # Running classifier on test set
    
    test_pred_label = model_k.predict(test_set)
    
    # Generating classification errors for both the training set and test set
    
    training_error_k = 1 - accuracy(training_label, training_pred_label)
    train_class_err.append(training_error_k)
    
    test_error_k = 1- accuracy(test_label,test_pred_label)
    test_class_err.append(test_error_k)
```

```python
# Loading both the training and test datasets without headers

training_set = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_Q7_train.csv", 
                          delimiter=',', skiprows=1)

test_set = np.loadtxt("/Users/Reinhard/IDS705_Principles_of_Machine_Learning/Assignment 2/A2_Q7_test.csv", 
                      delimiter=',', skiprows=1)

# Creating a scatterplot for the training dataset

%config InlineBackend.figure_format = 'retina'

plt.figure(figsize=(7,7))
plt.scatter(training_set[:,1], training_set[:,2], color='red')
plt.ylabel('y value')
plt.xlabel("x value")
plt.title('The Scatterplot of the Training Dataset', fontsize=14, fontweight ="bold" )

plt.show()
```

**(b)**

```python
# Estimating a linear regression model 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Separating the features and target for the training set and transforming them into 2D arrays 

x_train = training_set[:,1].reshape(-1, 1)
y_train = training_set[:,2].reshape(-1, 1)

# Using the scikit learn LinearRegression module to fit the data

model_baseline = LinearRegression().fit(x_train, y_train)
y_pred = model_baseline.predict(x_train)

# Presenting the equation, R squared and MSE 

print("Baseline Equation: \nY = %f + %f * X" % (model_baseline.intercept_, model_baseline.coef_))

print("R squared: %.2f" % r2_score(y_train, y_pred))

MSE = np.mean((y_train - y_pred)**2)
print("Mean Squared Error: %.2f" % MSE)
```

**(c)**

```python
# Finding a better model fit with non-linear transformation on predictors

from sklearn.preprocessing import PolynomialFeatures

# According to the shape of the above plot with 2 local optimized points, we guess that the predictors may 
# include terms like x^3

# Determining degree of polynomial

trans = PolynomialFeatures(degree = 3) 

# Transforming the data

x_train_trans = trans.fit_transform(x_train) 

# Using the scikit learn LinearRegression module to fit the data

model_trans = LinearRegression().fit(x_train_trans, y_train)
y_pred_trans = model_trans.predict(x_train_trans)

intercept = model_trans.intercept_
coef = model_trans.coef_

# Presenting the equation, R squared and MSE 

print("Estimated Equation with Predictors Transformation: \nY = %f + %f * X + %f * X^2 + %f * X^3" % 
      (intercept, coef[0,1], coef[0,2], coef[0,3]))

print("R squared: %.2f" % r2_score(y_train, y_pred_trans))

MSE_trans = np.mean((y_train - y_pred_trans)**2)
print("Mean Squared Error: %.2f" % MSE_trans)
```

**(d)**

```python
# plotting the original data, and the two curves representing your models

# Generating a background

plot_range = np.linspace(-4,6,50).reshape(-1,1)

# Creating the scatterplot

plt.figure(figsize=(7, 7))
plt.scatter(x_train, y_train, marker='.', color='lightblue')

# Plotting the linear Model

plt.plot(x_train, y_pred, color='purple', label = 'Baseline Model')

# Plotting the non-linear Model

plot_range_trans = trans.fit_transform(plot_range)
plt.plot(plot_range,model_trans.predict(plot_range_trans), color = 'red', label ='Polynomial Model')
plt.title('Fitting the Two Models to the Dataset', fontsize=14, fontweight ="bold")
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend(loc = 'best')

plt.show()
```

**(e)**

```python
# Loading the test set and separating it into features and target

x_test = test_set[:,1].reshape(-1, 1)
y_test = test_set[:,2].reshape(-1, 1)

# Applying the baseline model to the test set

y_pred_test = model_baseline.predict(x_test)

# Presenting the R squared and MSE

r2_score_test = r2_score(y_test, y_pred_test)
print("R squared when fitting the baseline model to test set: %.2f" % r2_score_test)

MSE_test = np.mean((y_test - y_pred_test)**2)
print("Mean Squared Error for test set: %.2f" % MSE_test)

print('-'*80)

# Transforming data

x_test_trans = trans.fit_transform(x_test) 

# Applying the polynomial model to the test set

y_pred_test_trans = model_trans.predict(x_test_trans)

# Presenting the R squared and MSE

r2_score_test_trans = r2_score(y_test, y_pred_test_trans)
print("R squared when fitting the polynomial model to test set: %.2f" % r2_score_test_trans)

MSE_test_trans = np.mean((y_test - y_pred_test_trans)**2)
print("Mean Squared Error: %.2f" % MSE_test_trans)

```

**(f)**

> Comparing the performance of the two models both in training dataset and test dataset, we see that the polynomial model perform better than the baseline model, since the former is able to capture the non-linear nature within the whole dataset more accurately than its peer can do.

**(g)**

> The predictive capability of the models would be seriously weakened, since we were assuming the training set and the test set have same distribution when we built the above models. 

> To be more specific, when the test data were significantly different from the training dataset, model evaluating metrics such as R squared and Mean Squared Error would perform worse on the test dataset than do in the training dataset, but this comparision bears no predictive capability in this context and also has nothing to do with model selection due to the violated assumption.
